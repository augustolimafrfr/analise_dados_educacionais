# ANÁLISE DE DADOS EDUCACIONAIS DO ENSINO SUPERIOR

### Descrição dos dados

O Censo da Educação Superior, conduzido anualmente pelo Inep, é a pesquisa mais abrangente do Brasil sobre instituições de educação superior (IES), seus cursos, alunos e docentes. Utilizando dados do Sistema e-MEC, o Censo coleta informações sobre infraestrutura, vagas, candidatos, matrículas, ingressantes, concluintes e docentes. Seu propósito é fornecer estatísticas confiáveis para compreender e monitorar o sistema brasileiro de educação superior, subsidiar o Ministério da Educação em atividades de acompanhamento, avaliação, expansão e melhoria da qualidade, além de disponibilizar dados para o cálculo de indicadores que fundamentem políticas públicas.

### Objetivo desse projeto

O objetivo desse projeto é realizar uma análise exploratória dos dados educacionais e responder algumas perguntas através de técnicas que aprendi durante meus estudo.

- Pergunta 3: Qual é a taxa de ingresso e conclusão de cursos?

- Pergunta 4: Existe uma correlação entre o desempenho acadêmico (quantidade de alunos que concluíram com sucesso) e a forma de ingresso (vestibular, ENEM, transferência, etc.)?

```python
dados_limpos.head()
```

```python
#AGRUPANDO PELO NOME DO CURSO A QUANTIDADE TOTAIS DE INGRESSANTES E CONCLUINTES EM 2020
taxa_conc_ing = dados_limpos[['NO_CINE_ROTULO', 'QT_ING', 'QT_CONC', 'ING_VESTIBULAR', 'ING_ENEM', 'ING_AVALIACAO_SERIADA', 'ING_SELECAO_SIMPLIFICA', 'ING_EGR', 'ING_VG_REMANESC', 'ING_VG_PROG_ESPECIAL', 'ING_OUTRA_FORMA']].groupby(by = 'NO_CINE_ROTULO').sum()

#EXCLUDINDO OS CURSOS QUE POSSUEM PELO MENOS MIL INGRESSANTES E MIL CONCLUINTES NO ANO DE 2020:
taxa_conc_ing = taxa_conc_ing[taxa_conc_ing['QT_ING'] > 1000]
taxa_conc_ing = taxa_conc_ing[taxa_conc_ing['QT_CONC'] > 1000]

#PERCENTUAL DE FORMANDOS EM 2020 EM RELAÇÃO AO NÚMERO DE INGRESSANTES PARA 2020:
taxa_conc_ing['taxa_conc_ing'] = round(taxa_conc_ing['QT_CONC']/taxa_conc_ing['QT_ING']*100, 2)

#SELECIONANDO O TOP 20 DESSA TABELA
taxa_conc_ing = taxa_conc_ing.sort_values(by = 'taxa_conc_ing', ascending=False)[0:20]
taxa_conc_ing
```

### TAXA DE INGRESSO E CONCLUSÃO 

```python
plt.figure(figsize=(10, 8))
taxa_conc_ing['taxa_conc_ing'].sort_values().plot(kind='barh', width = 0.8)
plt.title('Taxa de concluintes por ingressantes', size = 16)
plt.xlabel('Curso', size = 12)
plt.ylabel('Taxa', size = 12)
plt.show()
```

Resposta pergunta 3: Esses são os top 20 cursos com maiores taxas de conclusão/ingresso para o ano de 2020. Porém, é interessante salientar que estamos apenas comparando a quantidade de pessoas que concluiram o curso em 2020 com a que ingressaram em 2020. O interessante para saber a quantidade de pessoas que concluiram seu curso com sucesso seria comparar o ano atual de conclusão com o ano de ingresso de cada estudante, porém, não possuimos esses dados nesse dataset.

### Correlação entre as variáveis 

```python
#MATRIZ DE CORRELAÇÃO
correlacao = taxa_conc_ing[['taxa_conc_ing', 'ING_VESTIBULAR', 'ING_ENEM', 'ING_AVALIACAO_SERIADA', 'ING_SELECAO_SIMPLIFICA', 'ING_EGR', 'ING_VG_REMANESC', 'ING_VG_PROG_ESPECIAL', 'ING_OUTRA_FORMA']].corr()
correlacao
```

```python
# Cria um mapa de calor (heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Mapa de Calor - Correlação entre Variáveis', size=16)

# Exibe o gráfico
plt.show()
```

Resposta pergunta 4: Como pode-se observar no heatmap acima, a taxa de conclusão/ingresso não possui forte correlação com a maioria das outras variáveis que descrevem a forma de ingresso na universidade. Dentre elas, a que possui maior correlação (valor igual a 0.41) é ING_EGR que representa aqueles que terminaram uma graduação e entraram em outra de área correlata logo em seguida.

#### Número de alunos com deficiencia que ingressaram, matriculados e concluites

```python
qnt_def_ing = dados_limpos['ING_ALUNO_DEFICIENTE'].sum()
qnt_def_ing

qnt_def_mat = dados_limpos['MAT_ALUNO_DEFICIENTE'].sum()
qnt_def_mat

qnt_def_conc = dados_limpos['CONC_ALUNO_DEFICIENTE'].sum()
qnt_def_conc

print('Resposta pergunta 5:')
print(f'Alunos com deficiencia que ingressaram em algum curso em 2020: {qnt_def_ing}.')
print(f'Alunos com deficiencia que estão matriculados em algum curso em 2020: {qnt_def_mat}.')
print(f'Alunos com deficiencia que concluiram algum curso em 2020: {qnt_def_conc}.')
```

#### Relação de alunos deficientes por curso 

```python
#TABELA ORGANIZADA PELO NOME DO CURSO, QUANTIDADE DE ALUNOS MATRICULADAS COM DEFICIENCIA E QUANTIDADE TOTAL DE ALUNOS MATRICULADOS
cursos_def = dados_limpos[['NO_CINE_ROTULO', 'MAT_ALUNO_DEFICIENTE', 'QT_MAT']].groupby(by = 'NO_CINE_ROTULO').sum()

#CALCULANDO A PORCENTAGEM DE ALUNOS COM DEFICIENCIA ENTRE O TOTAL MATRICULADO
cursos_def['porcentagem'] = round(cursos_def['MAT_ALUNO_DEFICIENTE']/cursos_def['QT_MAT']*100, 2)

#SELECIONANDO O TOP 15 EM VALORES ABSOLUTOS
cursos_def_top15_abs = cursos_def.sort_values(by = 'MAT_ALUNO_DEFICIENTE', ascending = False)[0:15]

#SELECIONANDO O TOP 15 EM PORCENTAGEM
cursos_def_top15_per = cursos_def.sort_values(by = 'porcentagem', ascending = False)[0:15]
```

```python
cursos_def_top15_per
```

#### Gráfico com valores absolutos

```python
plt.figure(figsize=(10, 8))
cursos_def_top15_abs['MAT_ALUNO_DEFICIENTE'].sort_values().plot(kind='barh', width = 0.8)
plt.title('Quantidade de alunos com deficiência', size = 16)
plt.xlabel('Quantidade de alunos', size = 12)
plt.ylabel('Numero de alunos', size = 12)
plt.show()
```

####  Gráfico com valores em porcentagem

```python
plt.figure(figsize=(10, 8))
cursos_def_top15_per['porcentagem'].sort_values().plot(kind='barh', width = 0.8)
plt.title('Quantidade de alunos com deficiência em % do número de matriculados', size = 16)
plt.xlabel('Porcentagem', size = 12)
plt.ylabel('% do número de matriculados', size = 12)
plt.show()
```

Resposta pergunta 6: Em números absolutos Pedagogia e Direito são os cursos com mais alunos com deficiência, em terceiro lugar Administração, seguidos por Medicina, Psicologia e Contabilidade. Porém, quando se olha a proporção de alunos com deficiência entre os matriculados o curso de Letras com foco em LIBRAS (linguaguem dos sinais) é o com mais alunos com deficiência matriculados, seguido por outros três cursos de Letras.
